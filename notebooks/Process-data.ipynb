```python
# Import Libraries
import scanpy as sc 
import pandas as pd
import numpy as np
```

```python
adata = sc.read("data/PBMC.merged.h5ad")
adata # This prints all the attributes (columns) present in this object. We need to identify which one corresponds to cell type and batch (if available)
```

```python
data = adata.to_df() # Extracts gene expression matrix
data['batch'] = adata.obs['batch']
data['labels'] = adata.obs['Cell type']
```

```python
types = data['batch'].unique()
source_batch = data[data['batch'] == types[0]].drop(["batch"], axis=1)
target_batch = data[data['batch'] == types[1]].drop(["batch"], axis=1)
```

```python
source_batch.shape, target_batch.shape
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]=""
import pandas as pd
from jind import JindLib

lname = "labels"

train_mat = source_batch.drop(lname, axis=1)
train_labels = source_batch[lname]

test_mat = target_batch.drop(lname, axis=1)
test_labels = target_batch[lname]

path = "./my_results"

obj = JindLib(train_mat, train_labels, path=path)
mat = train_mat.values
mat_round = np.rint(mat)
error = np.mean(np.abs(mat - mat_round))
if error == 0:
    obj.preprocess(count_normalize=True, logt=True)

obj.dim_reduction(5000, 'Var')

train_config = {'val_frac': 0.2, 'seed': 0, 'batch_size': 128, 'cuda': False,
                'epochs': 15}

obj.train_classifier(config=train_config, cmat=True)
```

```python
predicted_label1, log1 = obj.evaluate(test_mat, test_labels, frac=0.05, name="testcfmt.pdf", return_log=True)
# Save object for later use
obj.to_pickle("JindLib_obj.pkl")
```

```python
train_config = {'seed': 0, 'batch_size': 128, 'cuda': False,
                'epochs': 15, 'gdecay': 1e-2, 'ddecay': 1e-3, 'maxcount': 7}
obj.remove_effect(train_mat, test_mat, train_config, test_labels)

predicted_label2, log2  = obj.evaluate(test_mat, test_labels, frac=0.05, name="testcfmtbr.pdf", test=True, return_log=True)
```

```python
train_config = {'val_frac': 0.1, 'seed': 0, 'batch_size': 32, 'cuda': False,
                'epochs': 10}
obj.ftune(test_mat, train_config)
predicted_label3, log3  = obj.evaluate(test_mat, test_labels, frac=0.05, name="testcfmtbrftune.pdf", test=True, return_log=True)
```
